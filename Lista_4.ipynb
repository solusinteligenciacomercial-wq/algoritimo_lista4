```python
"""1. Escreva uma função recursiva que calcule a soma dos dígitos de um número inteiro positivo.
Exemplo: 123 =>6"""

def soma_digitos (n):
    if n < 10:
        return n
    return (n % 10) + soma_digitos(n // 10)
print (f"soma dos digitos de 123: {soma_digitos(123)}")
```

```python
"""2. Crie uma função recursiva que conte quantos dígitos tem um número inteiro.
Exemplo: 9876 =>4"""

def contar_digitos(n):
    if n < 10:
        return 1
    return 1 + contar_digitos(n // 10)

print (f"Digitos em 9878: {contar_digitos(9876)}")
print (f"Digitos em 5: {contar_digitos(5)}")
```

```python
"""3. Implemente uma função recursiva que inverta uma string.
Exemplo: "recursao" =>"oasrucer"
"""
def inverter_string(s):
    if len(s) <=1:
        return s
    return s[-1] + inverter_string(s[:-1])
print(f'"algoritmo" invertido; {inverter_string("algoritimo")}')


```

```python
"""5. Crie uma função recursiva que calcule o produto de dois números usando apenas soma.
Exemplo: 3 × 4 =>12"""

def produto_soma(a, b):
    if b == 0:
        return 0
    return a + produto_soma(a, b -1)
print(f"Produto de 3 x 4: {produto_soma(3, 4)}")
print(f"Produto de 5 x 2: {produto_soma(5, 2)}")
```
